# Aquisitando dados:

Para esse modelo vamos manter exatamente a mesma estrutura feita por _Karpathy_, inclusive em seus dados de treino. Desde modo, precisamos aquisitar os dados de treino.

```python
# Baixando o conjunto de dados de treinamento
url = 'https://raw.githubusercontent.com/karpathy/char-rnn/master/data/tinyshakespeare/input.txt'
response = requests.get(url) # Fazendo uma requisição GET para o URL
data = response.text # Extraindo o texto da resposta
print(data[0:1000])  # Amostragem
```

```python
# Avaliando texto:
print(f'Tamanho do dataset em caracteres: {len(data)} chars')
print(f'Tamanho do dataset em palavras: {len(data.split())} palavras')
```
