```python
import os
import streamlit as st
from langchain.llms import CTransformers
from ctransformers import AutoModelForCausalLM
```

# Configurando o LLM:

```python
!wget https://huggingface.co/TheBloke/Llama-2-7B-Chat-GGML/resolve/main/llama-2-7b-chat.ggmlv3.q2_K.bin
```

```python
# Modelo:
llm = CTransformers(
    model="llama-2-7b-chat.ggmlv3.q2_K.bin",
    model_type="llama"
)
```

# Utilizando Flask:

```python
# Construção da Página:
from flask import Flask,render_template
import requests

# Implementação externa:
from google.colab.output import eval_js
link = eval_js('google.colab.kernel.proxyPort(5000)')
print(f'Click no link abaixo para acessar o livro:\n',link)
```

```python
# Instanciando a API:
app = Flask(__name__)

@app.route('/')
def index():
    return render_template('index.html')

@app.route('/generate_poem', methods=['POST'])
def generate_poem():
    if request.method == 'POST':
        content = request.form['content']
        result = llm.predict("write poem about " + content)
        return render_template('result.html', result=result)

app.run(host='0.0.0.0', debug=True, use_reloader=False)
```
